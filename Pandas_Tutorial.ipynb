```python
last_five_columns = diabetes_df.iloc[:, -5:]
print(last_five_columns)
```

```python
# removing a column
diabetes_df.drop(columns="SkinThickness", axis=1) #remove SkinThickness column
```

```python
# removing a row
diabetes_df.drop(index=0, axis=0)
```

```python
# if you need to remove a column you need to give axis = 1
# if you need to remove a row you need to give axis = 0
```

```python
# locating a row using the index value
diabetes_df.iloc[2]
```

```python
# locating a particular column
print(diabetes_df.iloc[:,0])  # first column
print(diabetes_df.iloc[:,1])  # second column
print(diabetes_df.iloc[:,2])  # third column
print(diabetes_df.iloc[:,-1]) # last column
```
